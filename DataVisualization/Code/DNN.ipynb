```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation
from keras.optimizers import SGD
import csv
import numpy as np
import cPickle


X_pos = cPickle.load(open('raw_data/pos_vects.pkl','rb'))
X_neg = cPickle.load(open('raw_data/neg_vects.pkl','rb'))
X_neu = cPickle.load(open('raw_data/neu_vects.pkl','rb'))
print len(X_pos)
print len(X_neg)
print len(X_neu)

X_pos_train = X_pos[0:275]
X_neg_train = X_neg[0:500]
X_neu_train = X_neu[0:250]
y_pos_train = [1]*len(X_pos_train) + [0]*len(X_neg_train) + [0]*len(X_neu_train)
y_neg_train = [0]*len(X_pos_train) + [1]*len(X_neg_train) + [0]*len(X_neu_train)
y_neu_train = [0]*len(X_pos_train) + [0]*len(X_neg_train ) + [1]*len(X_neu_train)
X_train = X_pos_train + X_neg_train + X_neu_train


X_pos_test = X_pos[275:325]
X_neg_test = X_neg[500:664]
X_neu_test = X_neu[250:299]
y_pos_test = [1]*len(X_pos_test) + [0]*len(X_neg_test) + [0]*len(X_neu_test)
y_neg_test = [0]*len(X_pos_test) + [1]*len(X_neg_test) + [0]*len(X_neu_test)
y_neu_test = [0]*len(X_pos_test) + [0]*len(X_neg_test ) + [1]*len(X_neu_test)
X_test = X_pos_test + X_neg_test + X_neu_test

print len(X_train)
print len(X_test)
```

```python
y_pos_train = np.array(y_pos_train)
y_neg_train = np.array(y_neg_train)
y_neu_train = np.array(y_neu_train)
y_pos_test = np.array(y_pos_test)
y_neg_test = np.array(y_neg_test)
y_neu_test = np.array(y_neu_test)
n_feature = len(X_train[0])
print n_feature
```

```python
model_pos = Sequential()
model_pos.add(Dense(output_dim=100, input_dim=n_feature, init='uniform', activation='relu'))
model_pos.add(Dropout(0.5))
model_pos.add(Dense(output_dim=50, activation='relu'))
model_pos.add(Dropout(0.5))
model_pos.add(Dense(1, activation='sigmoid'))
model_pos.compile(loss='mse', optimizer='rmsprop')
```

```python
from keras.models import model_from_json

pos_config_list = []
pos_train_error_list = []
pos_test_error_list = []
for i in xrange(1,500):
    model_pos.fit(X_train, y_pos_train, nb_epoch=1, batch_size=250, verbose=0)
    train_error = model_pos.evaluate(X_train, y_pos_train, batch_size=5, verbose=0)
    pos_train_error_list.append(train_error)
    print train_error
    test_error = model_pos.evaluate(X_test, y_pos_test, batch_size=5,verbose=0)
    pos_test_error_list.append(test_error)
    print test_error
    print '**************************************************************'
    weights = model_pos.get_weights()
    #weights = weights.flatten()
    weights_lst = []
    for weight in weights:
        #weights_lst += weight
        weight = weight.flatten().tolist()
        weights_lst += weight
    pos_config_list.append(weights_lst)
print len(pos_config_list)
```
